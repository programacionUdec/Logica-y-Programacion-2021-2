```python
import numpy as np


nombre=np.array( ['Diana','jose','daniel','manuela'])
peso=np.array([56.8,72.5,40.5,80.4])  #peso en kg
altura=np.array([1.70,1.55,1.48,1.72]) #altura en m
ID=np.arange(0,len(nombre))


IMC=peso/altura**2

print(nombre)
print(ID)
print(IMC)
```

```python
juntando=np.array( list(zip(ID,peso,altura,IMC) )  )

print(juntando)
```

```python
import matplotlib.pyplot as plt

plt.plot(peso,IMC,marker='o',label='IMC vs Peso',c='r')
plt.plot(peso,altura,marker='*',label='Altura vs Peso',c='g')
plt.title('Título Grafica')
plt.xlabel('nombre eje x')
plt.ylabel('nombre eje y')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.hist(peso,bins=3)
```

```python
x=np.arange(0,10)
y=x**2
y2=x**3

#print(x)
#print(y)

plt.plot(x,y,label='cuadratica',marker='o',c='b' )
plt.plot(x,y2,label='cubica',marker='X',c='k')
plt.legend()
plt.title('funciones')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#lectura y escritura de archivos

def estadisticos(a):
    a=np.array(a)
    print('Retorna el promedio, minimo, maximo de una lista de datos.','\n','La lista ingresada',a)
    prom=np.sum(a)/len(a)
    minimo=np.min(a)
    maximo=np.max(a)
    print("El promedio, minimio y maximo de su lista es:")
    return prom,minimo,maximo

estadisticos([1,2,3])
```

```python
h=estadisticos([10,2,5,6])
print(h)

```

```python
guardar=np.savetxt('./reporte.txt',h,fmt='%2.2f')
```
