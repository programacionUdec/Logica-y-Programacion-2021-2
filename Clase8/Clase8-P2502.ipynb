```python
import numpy as np

nombre=np.array( ['Diana','jose','daniel'])
peso=np.array([56.8,72.5,40.5])  #peso en kg
altura=np.array([1.70,1.55,1.48]) #altura en m
ID=np.arange(0,len(peso))

IMC=peso/altura**2
print(ID)
print(nombre)
print(IMC)
```

```python
juntando=np.array( list(zip(ID,peso,altura,IMC) )   )

print(juntando)
```

```python
import matplotlib.pyplot as plt

plt.plot(peso,IMC,label='imc vs peso',c='r', marker='o')
plt.plot(peso,altura,label='altura vs peso',c='g', marker='*')
plt.title('Estudio del IMC')
plt.xlabel('peso (kg)')
plt.ylabel('altura (m)')
plt.legend()
plt.grid()
plt.show()
```

```python
x=np.array(range(0,5))
y=x**2
y2=x**3

plt.plot(x,y,label='Cuadratica',marker='o',c='b')
plt.plot(x,y2,label='Cubica',marker='X',c='k')
plt.legend()
plt.title('funciones')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# lectura y escritura de archivo

def estadisticos(a):
    a=np.array(a)
    print('Retorna el promedio, minimo, maximo de una lista de datos.','\n','La lista ingresada',a)
    prom=np.sum(a)/len(a)
    minimo=np.min(a)
    maximo=np.max(a)
    print("El promedio, minimio y maximo de su lista es:")
    return prom,minimo,maximo
    

estadisticos([1,2,3])
```

```python
h=estadisticos([10,2,5])

print(h)
```

```python
guardar=np.savetxt('./reporte.txt',h,fmt='%2f')
```
